###Task 1

1. Determine the Top 3 most common cuisines in the dataset.

```python
Cuisines_counts = df['Cuisines'].str.split(',').explode().value_counts()
print(Cuisines_counts.head(3))
```

2 Calculate the percentage of restaurants that serve each of the top cuisines.

```python
Total_restaurant = len(df)
Restaurant_by_cuisines = (Cuisines_counts / Total_restaurant)*100
print(Restaurant_by_cuisines.head(10))
plt.figure(figsize = (10,6))
plt.bar(Restaurant_by_cuisines.head(10).index , Restaurant_by_cuisines.head(10).values ,color = "r")
plt.xlabel('Cuisines')
plt.ylabel('Percentage of Restaurants')
plt.title('Percentage of Restaurants by Cuisines')
plt.show()
```

###Task 2

3. Identify the city with the highest number of restaurants in the dataset.

```python
City_by_restaurants = df.groupby('City')['Restaurant Name'].count().sort_values(ascending = False)
print(City_by_restaurants.head(1))
```

4. Calculate the average rating for restaurants in each city.

```python
Avgrating_restaurant_by_eachcity = df.groupby('City')['Aggregate rating'].mean()
print(Avgrating_restaurant_by_eachcity)
```

5. Determine the city with the highest average rating.

```python
city_highest_rating = df.groupby('City')['Aggregate rating'].mean().idxmax()
print(city_highest_rating)
```

###Task 3

6. Create a histogram of bar chart to visualize the distribution of price ranges among the restaurants.

```python
pricerange_by_restaurants = df.groupby('Price range')['Restaurant Name'].count()
print(pricerange_by_restaurants)
```

```python
plt.figure.figsize = (8,5)
plt.bar(pricerange_by_restaurants.index, pricerange_by_restaurants.values)
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurans')
plt.title('Distribuation of Price ranges')
plt.show()


```

7. Calculate the percentage of restaurants in each price range category .

```python
restaurants_by_pricerange = pricerange_by_restaurants * 100 / Total_restaurant
print(restaurants_by_pricerange)
```

```python
plt.figure.figsize = (8,8)
plt.pie(restaurants_by_pricerange.values, labels = restaurants_by_pricerange.index, autopct = '%1.1f%%')
plt.title('Percentage of Restaurants by Price range ')
plt.show()
```
