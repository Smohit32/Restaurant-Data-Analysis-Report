###Task 1: Restaurant Ratings
1. Analyzed the distribution of aggregate
ratings and determined the most common
rating range.

```python
Common_rating_range = df['Aggregate rating'].value_counts()
print(Common_rating_range)
plt.figure(figsize =(7,5))
plt.bar(Common_rating_range.index , Common_rating_range.values , color = "r")
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Common Rating Range')
plt.show()
```

2. Calculated the average number of votes
received by restaurants.

```python
Avg_votes_by_restaurant = df['Votes'].mean()
print(Avg_votes_by_restaurant)
```

###Task 4 : Restaurant Chains
5. Identified if there are any restaurant chains
present in the dataset.

```python
Restaurant_chains = df['Restaurant Name'].value_counts()
print(Restaurant_chains[Restaurant_chains >1])
```

6. Analyzed the ratings and popularity of
different restaurant chains.

```python

chains_data = df[df['Restaurant Name'].isin(Restaurant_chains.index)]
chains_stats = chains_data.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum'}).reset_index()
chains_stats = chains_stats.sort_values(by=['Aggregate rating', 'Votes'], ascending=[False, False])
print(chains_stats.head(20))
```
